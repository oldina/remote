 1.2. Изучите стоимости недвижимости.

```python
plt.figure(figsize=(10,10))
plt.hist(df['price'])
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.title('Стоимость недвижимости')
plt.grid()
```

На рынке недвижимости преобладают квартиры с менее низкой стоимостью

1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(10,10))
plt.hist(df['sqft_living'])
plt.xlabel('Площадь')
plt.ylabel('Количество')
plt.title('Жилая квадратура')
plt.grid()
```

Больше всего квартир с жилой площадью от 18000 до 28000

1.4. Изучите распределение года постройки.

```python
plt.figure(figsize=(10,10))
plt.hist(df['yr_built'], bins=20)
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.title('Распределение года постройки')
plt.grid()
```

Больше всего домов построено в период с 2005 по 2010 год

```python
wf = df['waterfront'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(wf, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```


Домов с видом на набережную меньше 1%

2.2. Изучите распределение этажей домов

```python
f = df['floors'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(f, autopct='%1.1f%%', labels=f.index)
plt.title('Этажи')
```

Больше всего одноэтажных домов

2.3. Изучите распределение состояния домов

```python
cn = df['condition'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(cn, autopct='%1.1f%%', labels=cn.index)
plt.title('Состояние домов')
```

Больше всего домов в удовлитворительном состоянии

Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
table = df.corr(numeric_only=True)
table = np.round(table, 1)
table[np.abs(table) < 0.3] = 0
table
```

```python
plt.figure(figsize=(12, 9))
sns.heatmap(table, annot=True, linewidths=.5, cmap='coolwarm')
```

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
```

Чем больше ванных комнат в доме, тем выше стоимость

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
```

Чем больше жилая площадь дома, тем выше стоимость

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=5)
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Зависимость стоимости от этажности')
```

Самая большая стоимость у домов в 2,5 этажа

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=2)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Зависимость стоимости от этажности')
```

Наибольшую стоимость имеют дома с видом на "4"

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
```
